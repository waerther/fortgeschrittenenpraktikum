zeitlicher und räumlicher Kohärenz vertraut

Kohärenzgrad: Maß für die Interferenzfähigkeit zweier Quellen. Komplex. Es gilt 
$$
\gamma = \frac{\Gamma_12 (\tau)}{\sqrt{\Gamma_11 (0) \Gamma_22 (0)}}
$$
und




Zeitliche Kohärenz: Kann bei Amplitudenaupaltungs-Interferometer (Bsp: Michelson Morley) beobachtet werden, wenn die Strahlen wieder zusammengeführt werden.
\textit{Ist im Hecht komisch formuliert, ka ob richtig}: Interferenz liegt nur dann bei solchen Experimenten vor, wenn der Wegunterschied geringer ist als Kohärenzlänge (k-Länge: Die Länge, die der Wegunterschied sein darf, wobei die Kohärenz noch signifikant ist)

Räumliche Kohärenz: Maß für Korrelation der Amplituden zwischen den beiden Wellen. Nimmt normalerweise ab für steigende Distanzen.

4.
Wie kann der Kontrast (auch Sichtbarkeit) eines Interferometers definiert werden?

Der Kontrast (oder Sichtbarkeit) ist nach Michelson folgendermaßen definiert
$$
\Nu (r) = \frac{I_{max} - I_{min}}{I_{max} + I_{min}}
$$
